```python
import staintools
%matplotlib inline
```

```python
stain_augmentor = staintools.StainAugmentor(method='vahadane', sigma1=0.2, sigma2=0.2)
```

```python
i1 = staintools.read_image('./data/i1.png')
staintools.plot_image(i1, fig_size=6)
```

```python
stain_augmentor.fit(i1)
```

```python
augmented = [stain_augmentor.transform() for _ in range(9)]

stack = staintools.make_image_stack(augmented)
staintools.plot_image_stack(augmented, width=3)
```
